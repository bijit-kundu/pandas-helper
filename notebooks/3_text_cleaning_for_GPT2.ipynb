```python
import re
import os
```

```python
# function to clean text from file
def clean_text(text):
    # Regex pattern to match decimal numbers, single digits, and [In]/[Out]
    pattern = r'\d+\.\d+|\b\d\b|\[In\]|\[Out\]'
    
    #TODO
    # Remove (continues on next page) (continued from previous page)
    # sentences starting with the following
    
    # Use re.sub() to remove all matches from the text
    cleaned_text = re.sub(pattern, '', text)

    # Remove extra spaces and clean up punctuation
    cleaned_text = re.sub(r'\s*,\s*', ', ', cleaned_text)  # Fix commas
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)       # Remove extra spaces
    cleaned_text = re.sub(r'\(continues on next page\)|\(continued from previous page\)',
                          '',
                          cleaned_text)                    # Remove the strings (continues on next page) (continued from previous page)
    cleaned_text = re.sub(r'The following.*?[.!?](?=\s|$)',
                          '', cleaned_text)                # Remove sentences starting with "The following"
    cleaned_text = re.sub(r'\[\]', '', cleaned_text)       # Remove the pattern []
    
    # Strip leading/trailing whitespace
    cleaned_text = cleaned_text.strip()
    
    return cleaned_text
```

```python
full_text = ''

# Directory to process
directory_path = '../data/contents/full-text/full-text.txt'
output_path = '../data/contents/full-text/full-text.txt'

try:
    # Check if the path is a file (and not a directory)
    if os.path.isfile(directory_path):
        # Open and process the file
        with open(directory_path, 'r') as file:
            content = file.read()
            # Process the file content
            full_text += clean_text(content)
                
except Exception as e:
    print(f'Failed to read {directory_path}: {e}')
```

```python
with open(output_path, 'w', encoding='utf-8') as output_file:
        output_file.write(full_text)
```
